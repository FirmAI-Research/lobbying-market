```python
import pandas as pd
import yfinance as yf
```

```python
ticker = '^GSPC'
SP500 = yf.download(ticker, start='2013-01-01', end='2023-12-31')
```

```python
SP500.head(10)
```

```python
SP500 = SP500.loc[:,['Adj Close']]
SP500['Tomorrow'] = SP500['Adj Close'].shift(-1)
SP500['Return'] = (SP500['Tomorrow'] - SP500['Adj Close']) / SP500['Adj Close']
SP500['Cumulative Return'] = (1 + SP500['Return']).cumprod()
SP500.head()
```

```python
SP500.to_csv('SP500.csv')
```
